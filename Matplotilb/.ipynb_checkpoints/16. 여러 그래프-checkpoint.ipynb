# 16. 여러 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스  글자가 깨지는 현상을 해결
```

```python
import pandas as pd
df = pd.read_csv('../Pandas/score.txt', sep='\t')
df.set_index('지원번호', inplace=True)
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10)) # 2 x 2 에 해당하는 plot 들을 생성
# fig, axs(두 개 이상의 데이터를 듀플 형태로 받아오는 것)
# subplots(하위에 plot들을 몇 개 넣을 건지 크기 지정) =(세로 수량, 가로 수량)
fig.suptitle('여러 그래프 넣기')
# suptitle(여러 그래프의 전체 제목 지정 = 'super title')

# 첫 번째 그래프
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수') # axs(사용하려는 그래프 지정)= [그래프 x좌표, y좌표] ex)[0,0][0,1] 등
axs[0, 0].set_title('첫 번째 그래프') # 여러 그래프가 있을 때 'set_title' 이용해서 그래프 이름 설정 가능(그냥 title 사용 x)
axs[0, 0].legend(loc='upper center') # 범례
axs[0, 0].set(xlabel='이름', ylabel='점수') # 레이블명도 set(xlabel='', ylabel='') 방식으로 설정(그냥 xlabel, ylabel 금지)
axs[0, 0].set_facecolor('lightyellow') # 전면 색
axs[0, 0].grid(ls='--', lw=0.5)

# 두 번째 그래프
axs[0, 1].plot(df['이름'], df['수학'], label='수학')
axs[0, 1].plot(df['이름'], df['영어'], label='영어')
axs[0, 1].legend()

# 세 번째 그래프
axs[1, 0].barh(df['이름'], df['키'])

# 네 번째 그래프
axs[1, 1].plot(df['이름'], df['사회'], color='g', alpha=0.5)
```
