# 15. 산점도 그래프
두 변수의 상관관계 그래프를 보여주는 것

```python
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스  글자가 깨지는 현상을 해결
```

```python
import pandas as pd
df = pd.read_csv('../Pandas/score.txt', sep='\t')
df.set_index('지원번호', inplace=True)
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['영어'], df['수학']) # scatter(산점도 그래프) =(x축 데이터, y축 데이터)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000  # 0~1까지의 랜덤 숫자 rand(필요 랜덤 추점 갯수)
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes) # s(포인터 크기)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500 # 1학년 = 500, 2학년 =1,000, 3학년 =1,500
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
            # c(color지정)=color지정 기준, cmap(color 제공 받는 명령어) = 제공 받으려는 것
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5, orientation='horizontal')
# colorbar(값별 목례 컬러바 생성), ticks(컬러바에 값별 이름 생성), label(컬러바 이름), shrink(컬러바 크기 조정 0~1)
# orientation(컬러바 위치 설정) = 'horizontal'(그래프 아래 쪽)
```
