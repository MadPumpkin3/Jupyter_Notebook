# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
import openpyxl
file_name = '2021_연령별인구현황_8월.xlsx'
df_m = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B,E:Y')
# usecols = column을 기준으로 가지고 오고 싶은 column을 여러개 지정하는 것(엑셀 column기준으로 입력)
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 1,195,951 -> 1195951 (정수형)
# str.replace = 지정 단어를 원하는 단어로 변환(바꿀 단어, 바꾸고 싶은 단어) 단위 콤마를 빈칸으로 변환 but, int형으로 변환되는 것은 아님
# str.astype = 원하는 타입으로 변환(str=문자열, int=정수형, float=실수형)
df_m
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B,AB:AV')
df_w.head(3)
# 앞서 진행한 남자 데이터의 column과 명칭이 동일해서 '0~1세.1'로 자동 표현됨.
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns # 컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
df_w.head(3)
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 단위 : 천 명 , 남자 데이터가 보이게 '-'를 붙임(-df_m.iloc[0])
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2021년 대한민국 인구 피라미드')
plt.savefig('2021_인구피라미드.png', dpi=100)
plt.show()
```
